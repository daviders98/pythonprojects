```

%matplotlib inline
import pandas as pd
import geopandas
import matplotlib.pyplot as plt


countries = geopandas.read_file('data/ne_110m_admin_0_countries.shp')
!pip install --upgrade geopandas
!pip install --upgrade pyshp
!pip install --upgrade shapely
!pip install --upgrade descartes
import numpy as np
countries2 = geopandas.read_file('data/ne_110m_admin_0_countries.shp')
import pandas as pd 
location = "/content/drive/MyDrive/Colab Notebooks/Proyecto/titanic.xlsx"
df = pd.read_excel(location)
dfMale = df[['Gender','CountryHome']].loc[df['Gender']=='male']
dfMale = dfMale.value_counts()
dfMale = pd.DataFrame(dfMale,columns=['Passengers']) 
dfMale = dfMale.reset_index()
dfMale = dfMale.drop(columns='Gender')
dfMale['CountryHome'][1] = 'United Kingdom'
countries2['DATOS']=0
def putnumMale():
  indexes = []
  values = []
  #obtener valores
  for i in range(countries2['NAME_LONG'].size):
    for j in range(dfMale['CountryHome'].size):
      if countries2.at[i,'NAME_LONG'] == dfMale.at[j,'CountryHome']:
        indexes.append(i)
        values.append(dfMale['Passengers'][j])
  #acomodar valores
  k=0
  for i in indexes:
    countries2['DATOS'][i] = values[k]
    k+=1
putnumMale()
countries2
map = countries2.plot(column='DATOS',legend=True, figsize=(10, 10),categorical= True)
map.set_title('Male passengers', fontdict={'fontsize': 40}, loc='center')
dfFem = df[['Gender','CountryHome']].loc[df['Gender']=='female']
dfFem = dfFem.value_counts()
dfFem = pd.DataFrame(dfFem,columns=['Passengers']) 
dfFem = dfFem.reset_index()
dfFem = dfFem.drop(columns='Gender')
dfFem['CountryHome'][1] = 'United Kingdom'
countries2['DATOS']=0
def putnumFemale():
  indexes = []
  values = []
  #obtener valores
  for i in range(countries2['NAME_LONG'].size):
    for j in range(dfFem['CountryHome'].size):
      if countries2.at[i,'NAME_LONG'] == dfFem.at[j,'CountryHome']:
        indexes.append(i)
        values.append(dfFem['Passengers'][j])
  #acomodar valores
  k=0
  for i in indexes:
    countries2['DATOS'][i] = values[k]
    k+=1
putnumFemale()
map = countries2.plot(column='DATOS', legend=True, figsize=(10, 10),categorical= True)
map.set_title('Female passengers', fontdict={'fontsize': 40}, loc='center')
```
